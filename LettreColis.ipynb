```python
class Lettre: # Définition de notre classe Lettre

    def __init__(self, addeslettre, adexplettre,poidslettre,modelettre,expformlettre,prixTimbre=0):
        self.addeslettre = addeslettre
        self.adexplettre = adexplettre
        self.poidslettre = poidslettre
        self.modelettre = modelettre
        self.expformlettre = expformlettre
        self.prixTimbre = prixTimbre
    
    def calculTimbrelettre(self):
        if self.modelettre == "normal":
            if self.expformlettre == "A4":
                self.prixTimbre = (2.5 + 1) * (self.poidslettre /1000)
            elif self.expformlettre == "A3":
                self.prixTimbre = (3.5 + 1) * (self.poidslettre / 1000)
            else:
                self.prixTimbre = "Format non reconnu"
        elif self.modelettre == "express":
            if self.expformlettre == "A4":
                self.prixTimbre = ((2.5 + 1) * (self.poidslettre /1000))*2
            elif self.expformlettre == "A3":
                self.prixTimbre = ((3.5 + 1) * (self.poidslettre / 1000))*2
            else:
                self.prixTimbre = "Format non reconnu"
        else:
                self.prixTimbre="Mode d'expédition non reconnu"
    
    def ToString(self):
        self.calculTimbrelettre()
        print("Lettre:")
        print("Adresse destination:", self.addeslettre ,sep=" ")
        print("Adresse expediation:", self.adexplettre ,sep=" ")
        print("Poids:", self.poidslettre ,sep=" ")
        print("Mode:", self.modelettre ,sep=" ")
        print("Format:", self.expformlettre ,sep=" ")
        print("Prix du timbre:", self.prixTimbre ,sep=" ")
        

class Colis: # Définition de notre classe Colis

    def __init__(self, addescolis, adexpcolis, poidscolis,modecolis,volume,prixTimbre=0):
        self.addescolis = addescolis
        self.adexpcolis = adexpcolis
        self.poidscolis = poidscolis
        self.modecolis = modecolis
        self.volume = volume
        self.prixTimbre = prixTimbre
    
    def calculTimbreColis(self):
        if self.modecolis == "normal":
            self.prixTimbre = 0.25 * self.volume * self.poidscolis
        elif self.modecolis == "express":
            self.prixTimbre = (0.25 * self.volume * self.poidscolis)*2
        else:
            self.prixTimbre="Mode d'expédition non reconnu"
            
    def ToString(self):
        self.calculTimbreColis()
        print("Colis:")
        print("Adresse destination:", self.addescolis ,sep=" ")
        print("Adresse expediation:", self.adexpcolis ,sep=" ")
        print("Poids:", self.poidscolis ,sep=" ")
        print("Mode:", self.modecolis ,sep=" ")
        print("Volume:", self.volume ,sep=" ")
        print("Prix du timbre:", self.prixTimbre ,sep=" ")
        
        
        
L1=Lettre("Lille","Paris",80,"express","A3")
L1.ToString()
C1=Colis("Lille","Paris",3500,"express",2.25)

C1.ToString()


```
